## Genetic Algorithm

```python
best_paths = GeneticAlgo(pop_size = 500, mut_rate = 0.9, generations = 100).generate()[1]
best_distances = list(map(lambda path: path.distance, best_paths))
generations = range(len(best_paths))

plt.figure(figsize=(10, 6))
plt.plot(range(len(best_paths)), best_distances, marker='o', linestyle='-')
plt.xlabel('Geração')
plt.ylabel('Melhor Distância')
plt.title('Evolução da Melhor Distância ao Longo das Gerações')
plt.grid(True)
plt.show()
```
